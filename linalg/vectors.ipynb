### Vectors Theory

#### Vectors and Operation
A **vector** is a directed segment characterized by its magnitude and direction. Vectors are usually denoted as $\mathbf{a}, \mathbf{b}, \mathbf{c}$, or in coordinate form: $\vec{a} = (a_1, a_2)$.

Main operations:
- Addition of vectors: $\vec{a} + \vec{b} = (a_1 + b_1, a_2 + b_2)$
- Multiplication by a scalar: $k\vec{a} = (ka_1, ka_2)$
- Dot product: $\vec{a} \cdot \vec{b} = a_1b_1 + a_2b_2$

Geometrically:
- Vector addition can be represented as the construction of a parallelogram.
- Multiplying by a scalar changes the vector’s length, and if the number is negative, its direction.


Example code for visualizing vectors and operations:

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define vectors
a = np.array([2, 1])
b = np.array([1, 3])
k = 1.5

# Setup plot
plt.figure(figsize=(6,6))
plt.grid(True)
plt.axis('equal')

# Display the vectors
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='b')

# Sum of vectors
c = a + b
plt.quiver(0, 0, c[0], c[1], angles='xy', scale_units='xy', scale=1, color='g', label='a + b')

# Scalar multiplication
plt.quiver(0, 0, (k*a)[0], (k*a)[1], angles='xy', scale_units='xy', scale=1, color='orange', label='k * a')

plt.legend()
plt.title("Vector addition and scalar multiplication")
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.show()
```

#### Vector Spaces

A **vector space** (or **linear space**) $V$ over a field $F$ (commonly $\mathbb{R}$ or $\mathbb{C}$) is a non-empty set equipped with two operations:

1. **Vector addition:** $V \times V \rightarrow V$, for all $\mathbf{u}, \mathbf{v} \in V$, the sum $\mathbf{u} + \mathbf{v} \in V$.
2. **Scalar multiplication:** $F \times V \rightarrow V$, for all $a \in F, \mathbf{v} \in V$, the product $a \mathbf{v} \in V$.

These operations must satisfy the following axioms, for all $\mathbf{u}, \mathbf{v}, \mathbf{w} \in V$ and $a, b \in F$:

1. **Commutativity of addition:** $\mathbf{u} + \mathbf{v} = \mathbf{v} + \mathbf{u}$
2. **Associativity of addition:** $(\mathbf{u} + \mathbf{v}) + \mathbf{w} = \mathbf{u} + (\mathbf{v} + \mathbf{w})$
3. **Existence of zero vector:** There exists $\mathbf{0} \in V$ such that $\mathbf{v} + \mathbf{0} = \mathbf{v}$
4. **Existence of additive inverse:** For every $\mathbf{v} \in V$, there exists $-\mathbf{v} \in V$ such that $\mathbf{v} + (-\mathbf{v}) = \mathbf{0}$
5. **Compatibility of scalar multiplication with field multiplication:** $a(b\mathbf{v}) = (ab)\mathbf{v}$
6. **Identity element of scalar multiplication:** $1\mathbf{v} = \mathbf{v}$, where 1 is the multiplicative identity in $F$
7. **Distributivity of scalar multiplication with respect to vector addition:** $a(\mathbf{u} + \mathbf{v}) = a\mathbf{u} + a\mathbf{v}$
8. **Distributivity of scalar multiplication with respect to field addition:** $(a + b)\mathbf{v} = a\mathbf{v} + b\mathbf{v}$

**Examples of Vector Spaces**

1. $\mathbb{R}^n$ — the set of all n-tuples of real numbers forms a vector space over $\mathbb{R}$.
2. The set of all real-valued functions $f: \mathbb{R} \to \mathbb{R}$ is a vector space over $\mathbb{R}$.
3. The set of all polynomials of degree at most $n$ with real coefficients is a vector space over $\mathbb{R}$.

#### Hierarchy of Spaces

**Spaces** can be organized in a hierarchy, starting from the most basic — vector spaces — and progressing to more structured spaces such as normed spaces, inner product spaces, Banach spaces, and Hilbert spaces.

> **Every Hilbert space is a Banach space and every Banach space is a normed space, and both are vector spaces; but not every vector space is normed, and not every normed space admits an inner product.**

##### 1. Vector Spaces

A **vector space** $V$ over a field $F$ is defined as previously: a set with addition and scalar multiplication, satisfying the vector space axioms.

##### 2. Normed Linear Spaces

A **normed linear space** (or **normed vector space**) is a vector space $V$ equipped with a **norm**:

$$
\| \cdot \|: V \to \mathbb{R}
$$

such that for all $\mathbf{u}, \mathbf{v} \in V, a \in F$:

- $\| \mathbf{v} \| \geq 0$ (non-negativity; positivity only for nonzero vectors)
- $\| a \mathbf{v} \| = |a| \| \mathbf{v} \|$ (homogeneity)
- $\| \mathbf{u} + \mathbf{v} \| \leq \| \mathbf{u} \| + \| \mathbf{v} \|$ (triangle inequality)
- $\| \mathbf{v} \| = 0 \implies \mathbf{v} = 0$ (definiteness)

**Example:** $\mathbb{R}^n$ with Euclidean norm $\| \mathbf{v} \| = \sqrt{v_1^2 + \dots + v_n^2}$.

##### 3. Inner Product Spaces

An **inner product space** (or **pre-Hilbert space**) is a vector space $V$ endowed with an **inner product**:

$$
\langle \cdot, \cdot \rangle : V \times V \to F
$$

that satisfies, for all $\mathbf{u}, \mathbf{v}, \mathbf{w} \in V$ and $a \in F$:

- Linearity: $\langle a\mathbf{u} + \mathbf{v}, \mathbf{w} \rangle = a\langle \mathbf{u}, \mathbf{w} \rangle + \langle \mathbf{v}, \mathbf{w} \rangle$
- Symmetry: $\langle \mathbf{u}, \mathbf{v} \rangle = \overline{\langle \mathbf{v}, \mathbf{u} \rangle}$
- Positive-definiteness: $\langle \mathbf{v}, \mathbf{v} \rangle \geq 0$, equality if and only if $\mathbf{v} = 0$

The inner product induces a norm: $\| \mathbf{v} \| = \sqrt{\langle \mathbf{v}, \mathbf{v} \rangle}$.

**Example:** On $\mathbb{R}^n$, the standard dot product $\langle \mathbf{u}, \mathbf{v} \rangle = \sum_{i=1}^n u_i v_i$.

##### 4. Banach Spaces

A **Banach space** is a normed vector space $V$ that is **complete**; that is, every Cauchy sequence in $V$ converges to a point in $V$.

Completeness is a topological property, crucial for analysis and functional spaces.

**Example:** The space $l^p$ of all infinite sequences with finite $p$-norm.

##### 5. Hilbert Spaces

A **Hilbert space** is an inner product space that is also complete with respect to the norm induced by the inner product.

These spaces generalize Euclidean geometry to infinite-dimensional settings and form the backbone of functional analysis and quantum mechanics.

**Example:** The space $L^2$ of square-integrable functions.

### Norm

Let $V$ be a vector space over a field $F$ (typically, the real numbers $\mathbb{R}$ or complex numbers $\mathbb{C}$). A **norm** on $V$ is a function:

$$
\| \cdot \| : V \rightarrow [0, \infty)
$$

that assigns to every vector $\mathbf{v} \in V$ a non-negative real number $\| \mathbf{v} \|$ called the **norm** (or “length”) of $\mathbf{v}$, and satisfies the following three axioms for all $\mathbf{u}, \mathbf{v} \in V$, and all scalars $\alpha \in F$:

1. **Non-negativity (Positivity):** $\| \mathbf{v} \| \geq 0$, for all $\mathbf{v} \in V$. Moreover, $\| \mathbf{v} \| = 0$ if and only if $\mathbf{v} = \mathbf{0}$ (the zero vector).

2. **Absolute Scalability (Homogeneity):** $\| \alpha \mathbf{v} \| = |\alpha| \cdot \| \mathbf{v} \|$, for any scalar $\alpha \in F$ and vector $\mathbf{v} \in V$.

3. **Triangle Inequality (Subadditivity):** $\| \mathbf{u} + \mathbf{v} \| \leq \| \mathbf{u} \| + \| \mathbf{v} \|$, for all $\mathbf{u}, \mathbf{v} \in V$.


#### Classification and Types of Norms

p-Norms on $\mathbb{R}^n$ (Minkowski Norms):

- **$p=1$: Taxicab/Manhattan norm**
  $$
  \| \mathbf{x} \|_1 = \sum_{i=1}^{n} |x_i|
  $$

  - The Manhattan norm measures the total "city block" distance traveled along coordinate axes.
  - In two dimensions ($n=2$), it is the distance traversed moving strictly horizontally then vertically (as on a city’s rectangular grid), rather than "as the crow flies" (Euclidean norm).
- **$p=2$: Euclidean norm**
  $$
  \| \mathbf{x} \|_2 = \sqrt{ \sum_{i=1}^{n} x_i^2 }
  $$

  - The Euclidean norm generalizes the notion of the length of a vector in two- or three-dimensional space to arbitrary finite dimensions.
  - In $\mathbb{R}^2$, it is the familiar “straight-line” or “as-the-crow-flies” distance from the origin to the point $(x_1, x_2)$.
  - It is the norm induced by the standard inner (dot) product: 

  $$
  \langle \mathbf{x}, \mathbf{y} \rangle = \sum_{i=1}^{n} x_i y_i
  $$
  so that $\| \mathbf{x} \|_2 = \sqrt{\langle \mathbf{x}, \mathbf{x} \rangle}$.
  - **$p \to \infty$: Maximum/Uniform norm**
  $$
  \| \mathbf{x} \|_\infty = \max_{i=1, ..., n} |x_i|
  $$

  - The max norm measures the largest absolute entry in the vector.  
  - Geometrically, it represents the furthest distance along any coordinate axis from the origin.
  - In grid geometry, this is sometimes known as Chebyshev distance, the minimum number of moves in chess required to reach from one square to another with unrestricted king movement.

  #### Applications of Norms

- **Distance measurement:** Norms are used to compute distances in vector spaces; $d(\mathbf{u}, \mathbf{v}) = \|\mathbf{u} - \mathbf{v}\|$.
- **Optimization:** Constraints and objectives often use norms (e.g., $L^1$, $L^2$ regularization in machine learning).
- **Functional Analysis:** Norms are crucial for defining convergence, continuity, and completeness in infinite-dimensional spaces.

#### Some words about the unit ball

The unit ball in a normed vector space is the set of all vectors whose norm is less than or equal to one. It is defined as 
$$
B = \{ x \mid \|x\| \leq 1 \}.
$$
The boundary of the unit ball, the **unit sphere**, is the set of points with norm exactly one:
$$
S = \{ x \mid \|x\| = 1 \}.
$$

##### Properties
- The unit ball is always convex and symmetric about the origin because norms satisfy the triangle inequality and absolute scalability.
- In Euclidean spaces with the standard inner product, the unit ball is the familiar geometric ball (or disk in 2D) defined by $\|x\|_2 = \sqrt{x_1^2 + \cdots + x_n^2} \leq 1$.
- For other norms, such as $L^p$ norms, the shape of the unit ball changes and may become polyhedral or more complex in shape.

Examples:

- **Manhattan Norm ($p = 1$):** $|x_1| + |x_2| \leq 1$
  The unit ball is a **diamond** (square rotated by 45°). The points $(\pm1, 0)$ and $(0, \pm1)$ are the vertices.

- **Euclidean Norm ($p = 2$):** $x_1^2 + x_2^2 \leq 1$
  The unit ball is a **circle** centered at the origin with radius 1.

- **Max Norm ($p = \infty$):** $\max(|x_1|, |x_2|) \leq 1$
  The unit ball is a **square**, with sides parallel to axes, spanning from $(-1, -1)$ to $(1, 1)$.

- **Intermediate cases ($1 < p < \infty$):**
  The unit ball is a **smooth, centrally symmetric, convex shape**—it interpolates continuously between the diamond ($p=1$), circle ($p=2$), and square ($p=\infty$).

Usage examples:

1. **Normalization of Vectors:** In many algorithms, vectors are normalized to have length 1, meaning they lie on the unit ball's surface (the unit sphere). This ensures fair comparisons of direction rather than magnitude.

2. **Convex Optimization:** Constraints are often expressed as vectors lying inside a unit ball under some norm, such as $\|x\|_2 \leq 1$. This shapes the feasible region for optimization problems.



```python
x = np.array([2, -3, 1])

# L1 norm
l1 = np.linalg.norm(x, ord=1)

# L2 norm
l2 = np.linalg.norm(x, ord=2)

# Infinity norm
linf = np.linalg.norm(x, ord=np.inf)
```

```python
# For each angle θ, the radius r is computed so that ∥(r cos⁡(θ),r sin(⁡θ))∥p=1
def minkowski_unit_ball(p, num_points=400):
    theta = np.linspace(0, 2 * np.pi, num_points)
    # Parametric points on unit circle.
    # The cosine function gives the x-coordinate corresponding to the horizontal projection of an angle θ, 
    # while sine gives the y-coordinate corresponding to the vertical projection.
    x = np.cos(theta)
    y = np.sin(theta)
    
    # Normalize according to Lp norm: (|x|^p + |y|^p)^{1/p} = 1
    # Solve for radius r: r = 1 / (|cosθ|^p + |sinθ|^p)^{1/p}
    r = 1 / (np.abs(x)**p + np.abs(y)**p)**(1/p)
    
    # Scale unit circle points by radius r to place on unit ball boundary
    return r * x, r * y

p_values = [1, 1.5, 2, 4, 10, 50]

plt.figure(figsize=(6, 6))
for p in p_values:
    x, y = minkowski_unit_ball(p)
    plt.plot(x, y, label=f'p = {p}')
    
plt.title('Unit Balls for Minkowski Lp Norms in 2D')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.grid(True)
plt.show()
```

- For $p < 1$ (not a true norm), the unit ball becomes star-shaped, non-convex, and therefore loses norm properties.
- For **$p \in [1, \infty]$**:
  - $p = 1$: corners are sharp, edges are straight (diamond/octahedron/cross-polytope).
  - $p = 2$: corners are smoothest—circle/sphere.
  - $p \to \infty$: sides flatten, shape approaches a cube.
